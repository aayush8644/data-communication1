```python
# Example 1: Simple if
temperature = 25
print(f"Temperature: {temperature}°C")

if temperature > 30:
    print("It's hot outside! 🥵")
elif temperature > 20:
    print("It's warm outside! 😊")
elif temperature > 10:
    print("It's cool outside! 🧥")
else:
    print("It's cold outside! 🥶")
```
